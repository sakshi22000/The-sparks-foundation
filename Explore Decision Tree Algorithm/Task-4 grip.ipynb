# Task 4: Explore Decision Tree Algorithm


### In this task we have create Decision Tree Classifier and Visualize it graphically using the same iris dataset.

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_df.head()
```

```python
X = iris.data
y = iris.target
```

## Defining Decision Tree Algorithm

```python
# Creating our decision tree classifier
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(X, y)
```

## Visualize Decision Tree

```python
from sklearn import tree
plt.figure(figsize=(12,10))
tree.plot_tree(dtree, filled=True)
plt.title('Decision Tree')
```

## Splitting our Dataset

```python
from sklearn.model_selection import train_test_split
X_test, X_train, y_test, y_train = train_test_split(X, y, 
                                                    test_size = 0.2,
                                                    random_state = 0)
```

## Training our algorithm

```python
classifier = DecisionTreeClassifier(max_depth = 2, # preprune my tree to make sure it doesnt have a depth greater than 2 
                                    random_state = 0)
classifier.fit(X_train, y_train)
```

```python
classifier.predict(X_test[0:5])
```

## Measuring Performance

```python
score = classifier.score(X_test, y_test)
print(score)
```
